# Quality 
-only first 3 words in name 
-take care of nan values and remove 'kmpl/kmkg' in mileage to convert dtype to float 
-take care of missing values and remove 'CC' in engine 
-take care of missing values and remove 'bhp' in max_power 
-take care of missing values and remove units and make a seperate column for rpm in torque 
-take care of nan and missing values, convert values to int in seats


```python
cars=cars[~cars['engine'].isna()]
```

```python
cars.info()
```

```python
cars['name']=cars['name'].str.split(' ').str.slice(0,3).str.join(' ')
```

```python
cars.head()
```

```python
cars=cars.reset_index(drop=True)
```

```python
cars['mileage']=cars['mileage'].str.replace('kmpl',' ')
```

```python
cars['mileage']=cars['mileage'].str.replace('km/kg',' ').astype(float)
```

```python
cars['mileage'].head()
```

```python
cars['engine']=cars['engine'].str.replace('CC',' ').astype(float)
```

```python
cars['max_power'].unique()
```

```python
cars['max_power']=cars['max_power'].str.replace('bhp','')

```

```python
cars['seats']=cars['seats'].astype(int)
```

```python
cars.info()
```

```python
cars.head()
```

```python
cars['torque'].str.split('@')
```

```python
cars = cars.drop('torque',1)
```

```python
cars.info()
```

```python
cars['max_power'] = pd.to_numeric(cars['max_power'],errors = 'coerce')
```

```python
cars.info()
```

```python
cars[cars['max_power'].isnull()]
```

```python
cars['max_power'][4792]=cars['max_power'].mean()
```

```python
cars['max_power'].isnull().sum()
```

```python
cars.to_csv('cleaned_cars.csv')
```

## Model

```python
X = cars.drop(columns="selling_price")
```

```python
y = cars['selling_price']
```

```python
X
```

```python
y.shape
```

## Applying Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

## Creating an OneHotEncoder object to contain all the possible categories

```python
ohe=OneHotEncoder()
ohe.fit(X[['name','fuel','seller_type','transmission','owner']])
```

```python
ohe.categories_
```

```python
column_trans=make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','fuel','seller_type','transmission','owner']),
                                    remainder='passthrough')
```

## Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
RForest=RandomForestRegressor(n_estimators=5,random_state=329,min_samples_leaf=.0001)
```

```python
pipe=make_pipeline(column_trans,RForest)
```

```python
X_train.isnull().sum()
```

```python
X_train=X_train.reset_index(drop=True)
X_train
```

```python
y_train=y_train.reset_index(drop=True)
y_train
```

```python
y_test=y_test.reset_index(drop=True)
y_test
```

```python
X_test=X_test.reset_index(drop=True)
X_test
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred=pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
